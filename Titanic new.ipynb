```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df['Survived'].value_counts()
```

```python
df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Age'].median()
```

```python
df['Fare'].median()
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df = pd.get_dummies(df,drop_first=True)
df
```

```python
X = df.drop('Survived',axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=21)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
mdl = RandomForestClassifier(random_state=0)
mdl.fit(X_train,y_train)
ypred_train = mdl.predict(X_train)
ypred_test = mdl.predict(X_test)
from sklearn.metrics import accuracy_score
print('Training Accuracy:',accuracy_score(ypred_train,y_train))
print('Test Accuracy:',accuracy_score(ypred_test,y_test))
from sklearn.model_selection import cross_val_score
print('Cross Validation Score:',cross_val_score(mdl,X,y,cv=5).mean())
```

```python
from sklearn.model_selection import GridSearchCV
estimator =RandomForestClassifier(random_state=0)
param_grid = {'n_estimators':list(range(1,101))}
grid = GridSearchCV(estimator,param_grid,scoring='accuracy',cv=5)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
mdl = RandomForestClassifier(n_estimators=70,random_state=0)
mdl.fit(X_train,y_train)
ypred_train = mdl.predict(X_train)
ypred_test = mdl.predict(X_test)
from sklearn.metrics import accuracy_score
print('Training Accuracy:',accuracy_score(ypred_train,y_train))
print('Test Accuracy:',accuracy_score(ypred_test,y_test))
from sklearn.model_selection import cross_val_score
print('Cross Validation Score:',cross_val_score(mdl,X,y,cv=5).mean())
```
