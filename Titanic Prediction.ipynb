```python
from joblib import load
model = load('titanic.joblib')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('test.csv')
df
```

```python
df.info()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Age'].replace(np.nan,28,inplace=True)
df['Fare'].replace(np.nan,14.45,inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df
```

```python
df.drop(['Embarked_Q','Embarked_S'],axis=1,inplace=True)
```

```python
df
```

```python
prediction = model.predict(df)
```

```python
prediction
```

```python
pred = pd.DataFrame(prediction,columns=['Survived'],index=df['PassengerId'])
```
